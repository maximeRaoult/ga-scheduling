```python

num_generations=50

avg = [0]*(num_generations+1)

mutation_chances=[x / 100.0 for x in range(1,26,1)]
heights = [0]*(26)
for mutation_chance in mutation_chances:
    for fname in glob(f'results/trial4/*{mutation_chance}*'):
        results = np.load(fname)
        avg += results["best_solutions"]
    heights[int(mutation_chance*100)] =avg[-1]/10

fig, ax1 = plt.subplots(1,1)


ax1.bar(range(len(heights)), heights, color="red")
ax1.set_xticklabels(["0","0%","5%","10%","15%", "20%","25%"])
ax1.invert_yaxis()

ax1.set_xlabel("Mutation Chance")
ax1.set_ylabel("Fitness")
ax1.set_title("Final fitness of varying mutation\n chances, from 1% to 25%")


```

```python

num_generations=50
avg_normed = [0]*(num_generations+1)
avg_not_normed = [0]*(num_generations+1)

for fname in glob('results/trial7/*False,3,*'):
    results = np.load(fname)
    avg_normed += results["best_solutions"]#

for fname in glob('results/trial7/*False,1,*'):
    results = np.load(fname)
    avg_not_normed += results["best_solutions"]

fig, ax_left = plt.subplots()
ax_right = ax_left.twinx()

ax_left.set_yscale('symlog')
ax_left.set_ylabel("Cost (not normalised)", color="black")

ax_right.set_ylabel("Cost (normalised)", color="red", rotation=-90,labelpad=15)

ax_left.plot(avg_not_normed, color='black',label='Not normalised')
ax_right.plot(avg_normed, color='red',label='Normalised')
plt.title("Average fitnesses of normalised penalty functions\n vs not normalised penalty functions")

fig.legend(loc='lower right',bbox_to_anchor=(0.909, 0.11))


plt.show()
```
